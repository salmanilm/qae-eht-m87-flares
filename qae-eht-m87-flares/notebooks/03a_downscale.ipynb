```python
# downsample patches
import numpy as np1
from pathlib import Path
import shutil

src_dir = Path("../data/processed/patches")
dst_dir = Path("../data/processed/patches_4x4")
dst_dir.mkdir(exist_ok=True)

for patch_file in src_dir.glob("*.npy"):
    patch = np1.load(patch_file)
    
    # downsample: take average of 32×32 blocks
    if patch.shape == (128, 128):
        small = patch.reshape(4, 32, 4, 32).mean(axis=(1, 3))
    else:
        small = patch
    
    np1.save(dst_dir / patch_file.name, small)

print(f"Downsampled {len(list(dst_dir.glob('*.npy')))} patches to 4×4")
```

```python
# update split paths
import json
from pathlib import Path

split_dir = Path("../data/processed/splits")

for split_name in ["train", "val", "test"]:
    split_path = split_dir / f"{split_name}_files.json"
    
    with open(split_path, "r") as f:
        files = json.load(f)
    
    # Update paths
    for item in files:
        item["path"] = item["path"].replace("patches/", "patches_4x4/")
    
    # Save back
    with open(split_path, "w") as f:
        json.dump(files, f, indent=2)
    
    print(f"Updated {split_name}_files.json")

# verify
sample = json.load(open(split_dir / "train_files.json"))[0]
print(f"\nSample path: {sample['path']}")
```
